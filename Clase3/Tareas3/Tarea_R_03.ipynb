12.- Encuentra la clase, el tipo, el modo y el modo de almacenamiento de los siguientes valores: `Inf`, `NA`, `NaN`, `" "`.

```r
# Encontrando las clases:
class(Inf) # numeric
class(NA)  # logical
class(NaN) # numeric
class(" ") # character
```

```r
# Encontrando los tipos:
typeof(Inf)  # double
typeof(NA)   # logical
typeof(NaN)  # double
typeof(" ")  # character
```

```r
# Encontremos ahora los modos:
mode(Inf)  # numeric
mode(NA)   # logical
mode(NaN)  # numeric
mode(" ")  # character
```

```r
# Por último, encontremos los modos de almacenamiento:
storage.mode(Inf)  # double
storage.mode(NA)   # logical
storage.mode(NaN)  # double
storage.mode(" ")  # character
```

13.- Genera aleatoriamente `1000` mascotas , de las opciones `perro`, `gato`, `pollo` y `pez dorado`, con la misma probabilidad de que cada uno sea elegido. Muestra los primeros valores de la variable resultante y cuente el número de cada tipo de mascota.

```r
# Generamos aleatoriamente a las mascotas con la función sample()
(mascotas <- factor(sample(
  c("perro", "gato", "pollo", "pez dorado"), 
  1000, 
  replace = TRUE)))

# Recurrimos a head(para mostrar algunos elementos de nuestra variable, 
# además de sus Levels
head(mascotas)

# Cómo nuestra varible es un factor, es posible llamar a summary() para
# que nos muestre la cantidad de animales de cada tipo
summary(mascotas)
```

14.- Describe al menos dos maneras de asignar una variable.

```r
# Una manera poco convencional de asignar valores a una varible es mediante la función
# assign(), la cual para una local requiere de dos argumentos: el nombre de la variable
# a asignar y el valor que que le quieres dar.
assign("math", 0)
# Para una asignación global assign() toma un argumento extra:
assign("math", 0, globalenv())

# La manera más común de asignar valores a una variable es mediante "<-", esto 
# debido a razones históricas.
math <- 0
```
