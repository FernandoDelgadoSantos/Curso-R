(a) Sea $X$ el número de $\texttt{"unos"}$ obtenidos en `12` lanzamientos  de un dado. Entonces $X$ tiene una  distribución binomial $(n=12, p=1/3)$ . Calcule una tabla de probabilidades binomiales para $x=0, 1, \dots ,12$ por dos métodos:
 * Usando la fórmula para la densidad:
$$P(X = K)=\binom{n}{k}p^{k}(1 -p)^{n -k}$$
y la aritmética en R. Usa $\texttt{0:12}$ para la secuencia de $x$ valores y la función $\texttt{choose}$ para calcular los coeficientes binomiales $\binom{n}{k}$.

 * Usando la función $\texttt{dbinom}$ de **R** y comparar tus resultados con ambos métodos.

```r
# Definimos las siguientes variables
n = 12
p = 1/3
k = 0:12
# Calculamos la densidad de nuestra variable binomial usando choose()
choose(n,k)*(p**k)*((1 - p)**(n - k))
# Ahora usamos la función dbinom()
dbinom(k, n, p)
# Comparamos ambos resultados, primero solo viendo si son igualment cercanos con 
# all.equal(), y nos retorna TRUE pues los vvalores truncados son los mismos.
all.equal(dbinom(k, n, p), choose(n,k)*(p**k)*((1 - p)**(n - k)))

# Sin embargo, identical() nos retorna FALSE ya que los valores difieren al final 
# sus cifras decimales.
identical(dbinom(k, n, p), choose(n,k)*(p**k)*((1 - p)**(n - k)))
```

(b) Sea $X$ el número de $\texttt{"unos"}$ obtenidos en `12` lanzamientos  de un dado. Entonces $X$ tiene una  distribución binomial $(n=12, p=1/3)$. Calcule el CDF para $x=0, 1, \dots ,12$ por dos métodos:
 * Usando la función $\texttt{cumsum}$ y el resultado del ejercicio anterior.
 * Con el uso de la función $\texttt{pbinom}$. ¿Qué es $P(X >7)$?.

```r
# Independientemente de si usamos choose(n,k)*(p**k)*((1 - p)**(n - k)) o 
# dbinom(k, n, p) el último termino nos da 1 como debería ser en el cálculo 
# de la CDF.
cumsum(dbinom(k, n, p))
cumsum(choose(n,k)*(p**k)*((1 - p)**(n - k)))

# Igualmente con la función pbinom() 
# para el calculo del CDF con lo uqe el último termino resulta ser 1, como debería ser
pbinom(k, n, p)

# P(X > 7) nos brinda la probabilidad de uqe nuestra variable aleatoria sea mayor 
# a 7, y su valor esta dado por:
sum(dbinom(k, n, p)[9:13])
```
